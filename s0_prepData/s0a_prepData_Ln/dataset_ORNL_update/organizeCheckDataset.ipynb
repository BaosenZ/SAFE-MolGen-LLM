```python
import pandas as pd
```

```python
df_view = pd.read_excel('../data_template/data_template.xlsx', header=1)
df_view
```

```python
df_cleanData = pd.read_excel('output_cleaned_data.xlsx')
print(df_cleanData.shape)
df_cleanData.columns
```

```python
df_final = df_view
df_final['Entry_ID'] = [i for i in range(df_cleanData.shape[0])]
df_final['Extractants_count'] = [1] * df_cleanData.shape[0]
df_final['Comments'] = ['updated ORNL'] * df_cleanData.shape[0]
df_final.head(2)
```

```python
df_final['SMILES'] = df_cleanData['SMILES']
df_final['Extractant_conc_M'] = df_cleanData['Ligand_conc_M']
df_final['Solvent_A'] = df_cleanData['Solvent A']
df_final['Solvent_B'] = df_cleanData['Solvent B']

# Add oxidation state as III for all metal
df_final['Metal'] = df_cleanData['Metal']+"(III)"

df_final['Metal_conc_mM'] = df_cleanData['Ln conc. (mM)']

df_final['Acid_type'] = df_cleanData['Acid_type']
df_final['Acid_conc_M'] = df_cleanData['Acid conc. (M)']
df_final['Temperature_K'] = df_cleanData['Temperature_K']
df_final['Distribution_ratio'] = 10**df_cleanData['log D']
df_final['Log_D'] = df_cleanData['log D']
df_final['DOI'] = df_cleanData['DOI']

df_final['Volume_fraction_B'] = df_cleanData['Volume ratio of solvent B']/10
df_final['Volume_fraction_A'] = df_cleanData['Volume ratio of solvent A']/10

print(df_final.shape)
df_final.head(3)
```

```python
output_file = "output_db_Ln_ORNL.xlsx"
df_final.to_excel(output_file, index=False)
```
