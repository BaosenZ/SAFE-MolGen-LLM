```python
import pandas as pd
```

```python
df_template = pd.read_excel('../data_template/data_template.xlsx', header=1)
df_template
```

```python
df_cleanData = pd.read_excel('output_cleaned_data.xlsx')
print(df_cleanData.shape)
df_cleanData.columns
```

```python
df_cleanData.head(3)
```

```python
df_final = df_template.copy()
df_final['Entry_ID'] = [i for i in range(df_cleanData.shape[0])]
df_final['Extractants_count'] = [1] * df_cleanData.shape[0]
df_final['Comments'] = ['ACSEPT'] * df_cleanData.shape[0]
df_final.head(3)
```

```python
df_final['SMILES'] = df_cleanData['SMILES']
df_final['Extractant_conc_M'] = df_cleanData['extractant1 conc. (M)']
df_final['Solvent_A'] = df_cleanData['solvent1']
df_final['Solvent_B'] = df_cleanData['solvent2']

# By looking at the PDF report, all metals are III oxidation state
df_final['Metal'] = df_cleanData['metal'] + '(III)'

# The metal concentration is approximated as zero
df_final['Metal_conc_mM'] = [0] * df_cleanData.shape[0]

# By looking at the PDF report, all acids are HNO3
df_final['Acid_type'] = ['HNO3'] * df_cleanData.shape[0]

df_final['Acid_conc_M'] = df_cleanData['acid conc. (M)']
df_final['Temperature_K'] = df_cleanData['Temperature_K']
df_final['Distribution_ratio'] = 10**df_cleanData['logD']
df_final['Log_D'] = df_cleanData['logD']

df_final['DOI'] = ['https://cordis.europa.eu/project/id/211267'] * df_cleanData.shape[0]

df_final['Volume_fraction_B'] = df_cleanData['volume_fraction_B']
df_final['Volume_fraction_A'] = 1- df_cleanData['volume_fraction_B']

print(df_final.shape)
df_final.head(3)
```

### Output file

```python
output_file = "output_db_Ln_ACSEPT.xlsx"
df_final.to_excel(output_file, index=False)
```
