All datasets previously are integrated here **simply** by DOI to have a whole view of datasets. Another reason is to prepare the conditions, such as unify the solvent name, etc. **Except for solvent, acid, we do not use other inforamtion!**

```python
import pandas as pd
```

```python
df_ornl = pd.read_excel('../dataset_ORNL_update/output_db_Ln_ORNL.xlsx')
df_ornl.shape
```

```python
df_dga = pd.read_excel('../dataset_DGA/output_db_Ln_DGA.xlsx')
df_dga.shape
```

```python
# Get the list of DOIs from the first dataframe
doi_to_exclude = df_ornl['DOI'].unique()

# Filter the second dataframe to exclude these DOIs
df_filtered = df_dga[~df_dga['DOI'].isin(doi_to_exclude)]

# Integrate the dataframes
df_combined_ornl_dga = pd.concat([df_ornl, df_filtered], ignore_index=True)

df_combined_ornl_dga.shape
```

```python
df_combined_ornl_dga
```

```python
df_ideal = pd.read_excel('../dataset_IDEaL/output_db_Ln_IDEaL.xlsx')

df_ideal.shape
```

```python
df_ideal
```

```python
doi_to_exclude_2 = df_combined_ornl_dga['DOI'].unique()

df_filtered_2 = df_ideal[~df_ideal['DOI'].isin(doi_to_exclude_2)]

df_combined_ornl_dga_ideal = pd.concat([df_combined_ornl_dga, df_filtered_2], ignore_index=True)

df_combined_ornl_dga_ideal.shape
```

```python
df_acsept = pd.read_excel('../dataset_ACSEPT/output_db_Ln_ACSEPT.xlsx')

df_acsept.shape
```

```python
df_combined_ornl_dga_ideal_acsept = pd.concat([df_combined_ornl_dga_ideal, df_acsept], ignore_index=True)

df_combined_ornl_dga_ideal_acsept.shape
```

```python
df_combined_ornl_dga_ideal_acsept.to_excel("output_integ_data.xlsx", index=False)
```
