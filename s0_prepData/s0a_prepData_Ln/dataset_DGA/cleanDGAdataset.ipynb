```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('original/DGA_Literature_Review.xlsx', header=1)
print(df.shape)
df.head(3)
```

### Filter metals

```python
Ln_list = ['La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu']
metal_pattern = '|'.join(Ln_list)

metal_pattern
```

```python
df_filtered = df[df['Metal Identity'].str.contains(metal_pattern, na=False, regex=True)]

print(df_filtered.shape)
df_filtered.head(3)
```

### Remove Nan

```python
# If more than one extranctants, delete the corresponding row
df_filtered = df_filtered[~df_filtered['Name'].str.contains(';', na=False)]

print(df_filtered.shape)
df_filtered.head(3)
```

```python
# If more than one metal, delete the corresponding row (multiple metals can affect the D value)
df_filtered = df_filtered[~df_filtered['Aqueous Phase Metals'].str.contains(';', na=False)]

print(df_filtered.shape)
df_filtered.head(3)
```

```python
# Remove rows where value is NaN
df_filtered.loc[:, 'Temperature (C)'] = pd.to_numeric(df_filtered['Temperature (C)'], errors='coerce')
df_filtered = df_filtered[df_filtered['Temperature (C)'].notna()]

print(df_filtered.shape)
df_filtered.head(3)
```

```python
# Remove rows where 'Distribution Coefficient' is NaN or lower than or equal to 0
df_filtered['Distribution Coefficient'] = pd.to_numeric(df_filtered['Distribution Coefficient'], errors='coerce')
df_filtered = df_filtered[df_filtered['Distribution Coefficient'].notna() & df_filtered['Distribution Coefficient'] > 0]

print(df_filtered.shape)
```

### Fill Nan

```python
# If metal concentration is empty, replace with zero
df_filtered['Metal Concentration (mM)_cleaned'] = df_filtered['Metal Concentration (mM)']

df_filtered['Metal Concentration (mM)_cleaned'].fillna(0, inplace=True)
```

```python
df_filtered.loc[df_filtered['Metal Concentration (mM)_cleaned']=="Trace", 'Metal Concentration (mM)_cleaned'] = 0
print(df_filtered.shape)
```

### Calculation

```python
from rdkit import Chem
```

```python
def get_canonical_smiles(smiles):
    molecule = Chem.MolFromSmiles(smiles)
    return Chem.MolToSmiles(molecule, canonical=True)

df_filtered = df_filtered.rename(columns={'SMILES': 'SMILES_orig'})
df_filtered['SMILES'] = df_filtered['SMILES_orig'].apply(get_canonical_smiles)

print(df_filtered.shape)
```

```python
df_filtered['Temperature_K'] = df_filtered['Temperature (C)'] + 273.15

print(df_filtered.shape)
```

```python
# Calculate the base-10 logarithm and create a new column 'log_D'
df_filtered['log_D'] = np.log10(df_filtered['Distribution Coefficient'])

print(df_filtered.shape)
```

### Parse solvents

```python
import re
```

```python
# Parse the 'Solvent' column
def parse_solvent(solvent_str):
    if pd.isna(solvent_str):
        return pd.Series({'Solvent A': None, 'Solvent B': None, 'volume ratio': None})
    if 'with' in solvent_str:
        # Split at ' with '
        parts = solvent_str.split('with', 1)
        solvent_a = parts[0].strip()
        rest = parts[1].strip()
        # Use regex to extract volume ratio and Solvent B
        # "50 vol% solB"
        match1 = re.match(r'(\d+(?:\.\d+)?)\s*vol%\s*(.+)', rest)
        # "50 % solB"
        match2 = re.match(r'(\d+(?:\.\d+)?)\s*%\s*(.+)', rest)
        if match1:
            volume_ratio = match1.group(1)
            solvent_b = match1.group(2).strip()
            return pd.Series({'Solvent A': solvent_a, 'Solvent B': solvent_b, 'volume ratio': volume_ratio})
        elif match2:
            # print(match2)
            volume_ratio = match2.group(1)
            solvent_b = match2.group(2).strip()
            return pd.Series({'Solvent A': solvent_a, 'Solvent B': solvent_b, 'volume ratio': volume_ratio})
        else:
            volume_ratio = 0
            solvent_b = rest.strip()
        return pd.Series({'Solvent A': solvent_a, 'Solvent B': solvent_b, 'volume ratio': volume_ratio})
    else:
        solvent_a = solvent_str.strip()
        return pd.Series({'Solvent A': solvent_a, 'Solvent B': None, 'volume ratio': 0})

df_solvents = df_filtered['Solvent'].apply(parse_solvent)

# Concatenate the new columns with df_filtered
df_cleaned = pd.concat([df_filtered, df_solvents], axis=1)

df_cleaned['volume ratio'] = pd.to_numeric(df_cleaned['volume ratio'], errors='coerce')
df_cleaned['volume_fraction_B'] = df_cleaned['volume ratio']/100

df_cleaned.shape
```

```python
df_cleaned.to_excel('output_cleaned_data.xlsx', index=False)
```
