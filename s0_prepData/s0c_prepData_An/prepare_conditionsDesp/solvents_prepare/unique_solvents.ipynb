```python
import pandas as pd
```

```python
file_path = "../../integrate_datasets_simple/output_integ_data.xlsx"
df = pd.read_excel(file_path)
df.shape
```

```python
all_solvents_series = pd.concat([df['Solvent_A'], df['Solvent_B']])
all_solvents_series.shape
```

```python
# Remove duplicate SMILES strings
unique_solvents_series = all_solvents_series.drop_duplicates()
unique_solvents_series.shape
```

```python
unique_solvents_series.to_excel('output_unique_solvents.xlsx', sheet_name='solvents', index=False)
```
