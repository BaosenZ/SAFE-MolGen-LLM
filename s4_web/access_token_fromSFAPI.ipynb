```python
from authlib.integrations.requests_client import OAuth2Session
from authlib.oauth2.rfc7523 import PrivateKeyJWT
import requests

token_url= "https://oidc.nersc.gov/c2id/token"
client_id= "jgx54kcvavo4u"
private_key_path = "./priv_key.pem"

with open(private_key_path, "r") as file:
    private_key = file.read()

session = OAuth2Session(
    client_id, 
    private_key, 
    PrivateKeyJWT(token_url),
    grant_type="client_credentials",
    token_endpoint=token_url
)
fetch_token_result = session.fetch_token()
access_token = fetch_token_result['access_token'].strip()
print(access_token)
```

```python
r = requests.get("https://api.nersc.gov/api/v1.2/tasks?tags=%20", headers={ "accept": "application/json", "Authorization": f"Bearer {access_token}"})
print(r.json())
```
